```python
import cv2
from ultralytics import YOLO

# Загружаем модель YOLO
model = YOLO('../yolo-traffic-lights/traffic-lights-yolov8n.pt')

# Открываем видеофайл
cap = cv2.VideoCapture('../data/train_videos/akn00006_fqGg6dtL.mov')

# Проверяем, удалось ли открыть видео
if not cap.isOpened():
    print("Ошибка открытия видео")
    exit()

# Получаем параметры входного видео
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Настраиваем запись видео (используем кодек XVID)
out = cv2.VideoWriter('output_video.avi', cv2.VideoWriter_fourcc(*'XVID'), fps, (frame_width, frame_height))

while True:
    # Считываем кадр
    ret, frame = cap.read()
    if not ret:
        break  # Выход из цикла, если кадры закончились

    # Делаем предсказание на текущем кадре
    results = model(frame, conf=0.6)

    # Получаем кадр с предсказаниями
    marked_frame = results[0].plot()

    # Записываем обработанный кадр в выходной файл
    out.write(marked_frame)

# Освобождаем ресурсы
cap.release()
out.release()
cv2.destroyAllWindows()
```
