```python
from ultralytics import YOLO
import torch
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import wiener
import time
```

```python
model = YOLO('../yolo-traffic-lights/traffic-lights-yolov8n.pt')
```

```python
cap = cv2.VideoCapture('../data/train_videos/AKN00048.mp4')
```

```python
cap.set(cv2.CAP_PROP_POS_FRAMES, 1690)
```

```python
success, frame = cap.read()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6, 4))
plt.imshow(frame)

tstart = time.time()
marked_frame = model(frame, conf=0.3)
print(time.time() - tstart)

plt.figure(figsize=(6, 4))
plt.imshow(marked_frame[0].plot())
```
